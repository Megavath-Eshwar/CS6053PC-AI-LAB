```
# Monkey , Box and Banana [Room,window,door] (State-Space Search)
# Initial State : Location where it is at start
# Goal State : Grab Banana
# state = {Monkey : door; Box : window; Banana : center }

monkey = input("Enter Monkey location : ")
box = input("Enter Box location : ")
banana = input("Enter Banana location: ")
state = {"Monkey": monkey, "Box": box, "Banana": banana}

print("Initial State:", state)

if state["Monkey"]!= state["Box"]:
    print("Monkey moves towards the box.")
    state["Monkey"] = state["Box"]
    print("State:", state)

if state["Box"]= state["Banana"]:
    print("Monkey moves from box towards the banana.")
    state["Box"] = state["Banana"]
    state["Monkey"] = state["Banana"]
    print("State:", state)

print("Monkey grabs the Banana")
print("Goal State Reached!")
```
